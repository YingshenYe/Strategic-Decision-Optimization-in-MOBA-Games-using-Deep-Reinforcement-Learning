### 4.3 Theoretical Foundation
The Bellman equation forms the foundation of our approach:

$$Q(s,a) = r + \gamma \max_{a'} Q(s', a')$$

Where:
- $Q(s,a)$ = Expected cumulative reward for taking action $a$ in state $s$
- $r$ = Immediate reward
- $\gamma$ = Discount factor (0.95)
- $s'$ = Next state after action

**Double DQN Innovation:**
$$Q(s,a) = r + \gamma Q_{target}(s', \arg\max_{a'} Q_{online}(s', a'))$$

This decouples action selection from value estimation, reducing overestimation bias.

## **5. Problem Description**

### 5.1 Problem Formulation

We formulate the MOBA strategic decision-making problem as a **Markov Decision Process (MDP)**:

**States (S):** Each state represents the team’s condition at the 10-minute game snapshot, including:

- Economic condition: gold difference (Behind / Even / Ahead)
- Objective control: tower and dragon advantage
- Combat readiness: kill difference (Losing / Even / Winning)
- Vision, map pressure, and XP status: encoded as categorical or continuous features depending on the model, DQN uses a 22 dimensional one-hot encoding of these features, while DDQN compresses them into a 6 dimensional continuous state vector.

**Actions (A):** The agent selects from six strategic actions:

- **Tower_Push** – Apply lane pressure and threaten structures
- **Safe_Farm** – Prioritize gold/XP efficiency and reduce risk
- **Dragon_Control** – Contest or secure dragon objectives
- **Vision_Control** – Improve map visibility and deny enemy wards
- **Group_Attack** – Group to force coordinated engagements
- **Baron_Control** – Pressure or secure major neutral objectives

For the Q-Learning baseline, we restrict the action space to a simplified subset of three actions (Tower_Push, Safe_Farm, Dragon_Control) for interpretability and data efficiency.

**Rewards (R):**
The reward function balances multiple objectives:
```
reward = win_loss + economy_signal + objective_signal + action_specific_shaping
```
Reward components include:
- +1 / –1 for win or loss
- Small bonus for favorable gold or dragon differences
- Action-dependent shaping to encourage appropriate macro decisions

**Transitions (T):**
State transitions follow the order of match records:
- States evolve according to real game progression
- Actions do not alter future states
- Stochasticity arises from variability across recorded matches

### 5.2 Key Challenges

1. **High-dimensional state space** — Macro features span economy, objectives, vision, XP, and combat.
2. **Weak reward signals** — Early-game advantages translate inconsistently into final outcomes.
3. **Offline RL constraint** — Actions cannot affect future states, limiting exploration.
4. **Different model abstractions** — Tabular, one-hot, and continuous states reduce cross-model comparability.

## **6. Model Description**

### 6.1 Methods Overview

We implement three progressively sophisticated methods:

| Method | State Space | Function Approximation | Key Feature |
|--------|-------------|----------------------|-------------|
| Q-Learning | Discrete ([36] states) | Tabular | Exact values, interpretable |
| DQN | One-hot encoded([22] features) | Neural Network | Generalization |
| DDQN | Continuous ([6] features) | Dual Neural Networks | Reduced overestimation |

### 6.2 Q-Learning Architecture

**Q-Table Structure:**
- 36 states
  - Economy: Behind / Even / Ahead (3 levels)
  - Objective control: HardLosingObj / LosingObj / EvenObj / LeadingObj (4 levels)
  - Fight readiness: LosingFight / EvenFight / WinningFight (3 levels)
  - Total states = 3 × 4 × 3 = 36

- 3 actions:
  1. Tower_Push
  2. Safe_Farm
  3. Dragon_Control
- Total entries: 36 × 3 = 108 entries

**Update Rule:**
$$Q(s,a) \leftarrow Q(s,a) + \alpha [r + \gamma \max_{a'} Q(s',a') - Q(s,a)]$$

**Hyperparameters:**
- Learning rate (α): 0.1
- Discount factor (γ): 0.95
- Exploration (ε): 1.0 → 0.01 (decay: 0.995)

### 6.3 DQN Architecture

**Neural Network:**
```
Input Layer: 22 features
Hidden Layer 1: 128 neurons (ReLU)
Hidden Layer 2: 128 neurons (ReLU)
Hidden Layer 3: 64 neurons (ReLU)
Output Layer: 6 Q-values
```

**Key Components:**
- **Experience Replay Buffer:** 20,000 transitions
- **Batch Size:** 64
- **Optimizer:** Adam (lr= 0.001)
- **Loss Function:** Mean Squared Error

### 6.4 Double DQN Architecture

**Innovation:** Uses two networks to decouple action selection from value estimation

**Online Network:** Selects best action
$$a^* = \arg\max_{a'} Q_{online}(s', a')$$

**Target Network:** Evaluates selected action
$$Q_{target}(s', a^*)$$

**Update Frequency:** Target network updated every 5 episodes

### 6.5 Why These Methods?

1. **Q-Learning:** Provides an interpretable baseline to evaluate whether simple discrete macro-state representations are sufficient for strategic decision-making.
2. **DQN:** Handles high-dimensional one-hot state vectors and learns generalized value estimates beyond tabular limitations.
3. **DDQN:** Reduces overestimation and improves stability when learning from continuous state features and noisy reward signals.

### 6.6 Limitations and Future Improvements

**Current Limitations:**
- Weak reward signals make it difficult for DQN and DDQN to learn stable value estimates.
- State representations lack temporal context, preventing the models from capturing evolving game momentum.
- Action space mismatch across methods limits the fairness and comparability of performance results.

**Future Improvements:**
- Design richer and more consistent reward shaping to give the agent clearer learning signals.
- Introduce sequential features or recurrent architectures to model multi-step strategic dependencies.
- Unify and refine the action space and state representation to ensure fair cross-model comparisons and stronger generalization.

### 3. Define Environment for Q-Learning

````python
# State

def economy_simple(row):
    g = row["blueGoldDiff"]
    if g <= -2000: return "Behind"
    elif g >= 1500: return "Ahead"
    else: return "Even"

def objective_simple(row):
    # Towers are worth more tempo → multiply by 2
    tower_diff  = row["blueTowersDestroyed"] - row["redTowersDestroyed"]
    dragon_diff = row["blueDragons"] - row["redDragons"]

    d = dragon_diff + tower_diff * 2

    if d <= -2:
        return "HardLosingObj"
    elif d == -1:
        return "LosingObj"
    elif d == 0:
        return "EvenObj"
    else:  # d >= 1
        return "LeadingObj"

def fight_simple(row):
    kill_diff = row["blueKills"] - row["redKills"]
    if kill_diff <-3: return "LosingFight"
    elif kill_diff > 3: return "WinningFight"
    else: return "EvenFight"

def get_state_simple(row):
    return (
        economy_simple(row),
        objective_simple(row),
        fight_simple(row)
    )

# Define Actions (strategic focuses)
actions = ["Tower_Push", "Safe_Farm", "Dragon_Control"]
````

````python

# Shaped Reward: win/loss + small shaping on gold / dragon
def get_shaped_reward(row):
    reward = 1 if row["blueWins"] == 1 else -1
    reward += 0.1 * np.sign(row["blueGoldDiff"])
    reward += 0.2 * np.sign(row["blueDragons"] - row["redDragons"])
    return reward
````

### 7.2 Q learning Environment Implementation



````python
economy_map = {"Behind": 0, "Even": 1, "Ahead": 2}
objective_map = {"HardLosingObj": 0, "LosingObj": 1, "EvenObj": 2, "LeadingObj": 3}
fight_map = {"LosingFight": 0, "EvenFight": 1, "WinningFight": 2}
````

````python
class QEnv:
    def __init__(self, df):
        self.data = df.reset_index(drop=True)
        self.n_samples = len(df)
        self.current_step = 0

    def get_discrete_state(self, idx):
        row = self.data.iloc[idx]
        e, o, f = get_state_simple(row)

        e_idx = economy_map[e]
        o_idx = objective_map[o]
        f_idx = fight_map[f]

        # 36 states
        return e_idx * 12 + o_idx * 3 + f_idx

    def calculate_reward(self, idx):
        return get_shaped_reward(self.data.iloc[idx])

    def step(self, action_idx):
        reward = self.calculate_reward(self.current_step)

        self.current_step += 1
        done = self.current_step >= self.n_samples - 1

        if done:
            next_state = 0
        else:
            next_state = self.get_discrete_state(self.current_step)

        return next_state, reward, done

    def reset(self, start_idx=0):
        self.current_step = start_idx
        return self.get_discrete_state(start_idx)
````

### 7.3 Q-Learning Implementation (Baseline)


````python
class QLearningAgent:
    def __init__(self, n_states=36, n_actions=3, learning_rate=0.1, gamma=0.95, epsilon=1.0):
        self.n_states = n_states
        self.n_actions = n_actions
        self.lr = learning_rate
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995

        self.q_table = np.zeros((n_states, n_actions))

    def select_action(self, state, training=True):
        if training and random.random() < self.epsilon:
            return random.randint(0, self.n_actions - 1)
        return np.argmax(self.q_table[state])

    def update(self, state, action, reward, next_state, done):
        current_q = self.q_table[state, action]
        target = reward if done else reward + self.gamma * np.max(self.q_table[next_state])
        self.q_table[state, action] = current_q + self.lr * (target - current_q)

    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)
````

### 7.4 Deep Q-Network (DQN) Implementation

````python
class MOBAEnvironment:
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
        self.n_samples = len(df)
        self.current_step = 0

        # Your 6 states
        self.econ_vals   = ["Econ_HardBehind","Econ_Behind","Econ_Even","Econ_Ahead","Econ_Crushing"]
        self.xp_vals     = ["XP_HardBehind","XP_Behind","XP_Even","XP_Ahead","XP_Dominating"]
        self.tower_vals  = ["TowerBehind","TowerEven","TowerAhead"]
        self.dragon_vals = ["DragonBehind","DragonEven","DragonAhead"]
        self.vision_vals = ["LowVision","MedVision","HighVision"]
        self.combat_vals = ["LosingFight","EvenFight","WinningFight"]

        # Build mapping dicts
        self.state_maps = {
            "econ":   {v:i for i,v in enumerate(self.econ_vals)},
            "xp":     {v:i for i,v in enumerate(self.xp_vals)},
            "tower":  {v:i for i,v in enumerate(self.tower_vals)},
            "dragon": {v:i for i,v in enumerate(self.dragon_vals)},
            "vision": {v:i for i,v in enumerate(self.vision_vals)},
            "combat": {v:i for i,v in enumerate(self.combat_vals)}
        }

        # Total state vector length
        self.state_size = (
            len(self.econ_vals) +
            len(self.xp_vals) +
            len(self.tower_vals) +
            len(self.dragon_vals) +
            len(self.vision_vals) +
            len(self.combat_vals)
        )

        # Your actions
        self.actions = [
            "Tower_Push",
            "Safe_Farm",
            "Dragon_Control",
            "Vision_Control",
            "Group_Attack",
            "Baron_Control"
        ]
        self.n_actions = len(self.actions)

        print(f"Environment initialized!")
        print(f" - Dataset size: {self.n_samples}")
        print(f" - Action size: {self.n_actions}")
        print(f" - State vector length: {self.state_size}")

    # ----------------------------
    # Encoding State → Numeric Vector
    # ----------------------------
    def encode_state(self, row):
        econ   = econ_state(row)
        xp     = xp_state(row)
        tower  = tower_state(row)
        dragon = dragon_state(row)
        vision = vision_state(row)
        combat = combat_state(row)

        # One-hot encoding
        vec = []

        vec += tf.keras.utils.to_categorical(self.state_maps["econ"][econ], len(self.econ_vals)).tolist()
        vec += tf.keras.utils.to_categorical(self.state_maps["xp"][xp], len(self.xp_vals)).tolist()
        vec += tf.keras.utils.to_categorical(self.state_maps["tower"][tower], len(self.tower_vals)).tolist()
        vec += tf.keras.utils.to_categorical(self.state_maps["dragon"][dragon], len(self.dragon_vals)).tolist()
        vec += tf.keras.utils.to_categorical(self.state_maps["vision"][vision], len(self.vision_vals)).tolist()
        vec += tf.keras.utils.to_categorical(self.state_maps["combat"][combat], len(self.combat_vals)).tolist()

        return np.array(vec, dtype=np.float32)

    # -----------------------
    # Reward Function
    # -----------------------
    def compute_reward(self, row, action_idx):
        action = self.actions[action_idx]
        return get_reward(row, action)

    # -----------------------
    # Step
    # -----------------------
    def step(self, action_idx):
        row = self.df.iloc[self.current_step]
        reward = self.compute_reward(row, action_idx)

        self.current_step += 1
        done = self.current_step >= self.n_samples - 1

        if done:
            next_state = np.zeros(self.state_size, dtype=np.float32)
        else:
            next_row = self.df.iloc[self.current_step]
            next_state = self.encode_state(next_row)

        return next_state, reward, done

    def reset(self, start_idx=None):
        if start_idx is None:
            self.current_step = 0
        else:
            self.current_step = start_idx

        row = self.df.iloc[self.current_step]
        return self.encode_state(row)
````

````python
# Cell 3: Replay Buffer

class ReplayBuffer:
    def __init__(self, max_size=10000):
        self.buffer = deque(maxlen=max_size)

    def add(self, state, action, reward, next_state):
        experience = (state, action, reward, next_state)
        self.buffer.append(experience)

    def sample(self, batch_size):
        actual_size = min(batch_size, len(self.buffer))
        return random.sample(list(self.buffer), actual_size)

    def __len__(self):
        return len(self.buffer)

    def is_ready(self, batch_size):
        return len(self.buffer) >= batch_size

# Quick sanity check
test_buffer = ReplayBuffer(max_size=100)
for i in range(10):
    s = get_state(data.iloc[i])
    ns = get_state(data.iloc[(i+1) % len(data)])
    test_buffer.add(s, random.choice(actions), i, ns)

print("ReplayBuffer size:", len(test_buffer))
print("Sample:", test_buffer.sample(2))
````

````python
class DQNAgent:
    def __init__(self, state_size, n_actions, learning_rate=0.001):
        self.state_size = state_size
        self.n_actions = n_actions

        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.batch_size = 64

        self.buffer = ReplayBuffer(max_size=20000)
        self.model = self._build_model(learning_rate)

        print("DQN initialized!")
        print(f" - State size: {state_size}")
        print(f" - Actions: {n_actions}")

    def _build_model(self, lr):
        model = Sequential([
            Dense(128, activation="relu", input_shape=(self.state_size,)),
            Dense(128, activation="relu"),
            Dense(64, activation="relu"),
            Dense(self.n_actions, activation="linear")
        ])
        model.compile(optimizer=Adam(learning_rate=lr), loss="mse")
        return model

    def select_action(self, state):
        if np.random.random() < self.epsilon:
            return np.random.randint(self.n_actions)
        q = self.model.predict(state.reshape(1, -1), verbose=0)
        return np.argmax(q[0])

    def remember(self, s, a, r, s2, done):
        self.buffer.add(s, a, r, s2, done)

    def train(self):
        if len(self.buffer) < self.batch_size:
            return 0

        batch = self.buffer.sample(self.batch_size)

        states  = np.array([b[0] for b in batch])
        actions = np.array([b[1] for b in batch])
        rewards = np.array([b[2] for b in batch])
        next_states = np.array([b[3] for b in batch])
        dones = np.array([b[4] for b in batch])

        q_vals = self.model.predict(states, verbose=0)
        next_q = self.model.predict(next_states, verbose=0)

        for i in range(len(batch)):
            target = rewards[i] if dones[i] else rewards[i] + self.gamma * np.max(next_q[i])
            q_vals[i][actions[i]] = target

        loss = self.model.train_on_batch(states, q_vals)
        return loss

    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)
````

````python
class MOBAEnv:
    """
    Environment for LoL 10-min MOBA RL agent
    Fully compatible with your DQN/DDQN implementation.
    """

    def __init__(self, df):
        self.df = df.reset_index(drop=True)
        self.max_step = len(df) - 1

    def reset(self, start_idx=0):
        """
        Reset environment to a specific index.
        Returns:
            - state_discrete (unused, kept for compatibility)
            - state_continuous (6-dim vector)
        """
        self.current_step = start_idx

        row = self.df.iloc[self.current_step]
        state = state_to_vector(get_state(row))

        return None, state  # discrete_state unused

    def step(self, action_idx):
        """
        Take an action and move to next state.
        Returns:
            next_state_discrete, next_state_continuous, reward, done
        """

        # Map action index → action name
        action = actions[action_idx]

        # Current and next row
        row = self.df.iloc[self.current_step]
        next_row = self.df.iloc[min(self.current_step + 1, self.max_step)]

        # Reward from current state & action
        reward = get_reward(row, action)

        # Next state
        next_state_continuous = state_to_vector(get_state(next_row))

        # Termination condition
        self.current_step += 1
        done = self.current_step >= self.max_step

        return None, next_state_continuous, reward, done
````

````python
class ReplayBuffer:
    def __init__(self, max_size=10000):
        self.buffer = deque(maxlen=max_size)

    # Now accept (state, action, reward, next_state, done)
    def add(self, state, action, reward, next_state, done):
        experience = (state, action, reward, next_state, done)
        self.buffer.append(experience)

    def sample(self, batch_size):
        actual_size = min(batch_size, len(self.buffer))
        return random.sample(self.buffer, actual_size)

    def __len__(self):
        return len(self.buffer)
````

````python
class DDQNAgent:
    """
    Double Deep Q-Network Agent

    Key Innovation: Decouples action selection from Q-value estimation
    - Online network: Selects best action
    - Target network: Evaluates selected action

    This reduces overestimation bias common in standard DQN
    """

    def __init__(self, state_size=6, n_actions=6, learning_rate=0.001):
        self.state_size = state_size
        self.n_actions = n_actions
        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.98
        self.learning_rate = learning_rate
        self.batch_size = 128
        self.update_target_freq = 5  # Episodes

        # Build dual networks
        self.online_network = self._build_model()
        self.target_network = self._build_model()
        self.update_target_network()  # Initialize target with online weights

        # Experience replay
        self.buffer = ReplayBuffer(max_size=10000)

        print("Double DQN Agent initialized!")
        print(f"  - State size: {state_size}")
        print(f"  - Actions: {n_actions}")
        print(f"  - Target update frequency: Every {self.update_target_freq} episodes")

    def _build_model(self):
        """Build neural network"""
        model = Sequential([
            Dense(64, activation='relu', input_shape=(self.state_size,)),
            Dense(128, activation='relu'),
            Dense(32, activation='relu'),
            Dense(self.n_actions, activation='linear')
        ])
        model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss='mse')
        return model

    def update_target_network(self):
        """Copy weights from online to target network"""
        self.target_network.set_weights(self.online_network.get_weights())

    def select_action(self, state, training=True):
        """Epsilon-greedy action selection using online network"""
        if training and np.random.random() < self.epsilon:
            return np.random.randint(self.n_actions)

        state = np.array(state).reshape(1, -1)
        q_values = self.online_network.predict(state, verbose=0)
        return np.argmax(q_values[0])

    def remember(self, state, action, reward, next_state, done):
        self.buffer.add(state, action, reward, next_state, done)

    def train(self):
        """
        Double DQN Training:
        1. Online network selects best action: a* = argmax Q_online(s', a')
        2. Target network evaluates: Q_target(s', a*)
        """
        if len(self.buffer) < self.batch_size:
            return 0

        batch = self.buffer.sample(self.batch_size)
        states = np.array([exp[0] for exp in batch])
        next_states = np.array([exp[3] for exp in batch])

        current_q = self.online_network.predict(states, verbose=0)
        next_q_online = self.online_network.predict(next_states, verbose=0)
        next_q_target = self.target_network.predict(next_states, verbose=0)

        for i, (state, action, reward, next_state, done) in enumerate(batch):
            if done:
                target = reward
            else:
                best_action = np.argmax(next_q_online[i])
                target = reward + self.gamma * next_q_target[i][best_action]

            current_q[i][action] = target

        loss = self.online_network.train_on_batch(states, current_q)
        return loss

    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

# Initialize DDQN agent
ddqn_agent = DDQNAgent(state_size=6, n_actions=6)
````

- Develop a continuous learning pipeline that updates models as new match data becomes available.
- Explore multi agent extensions to model coordinated team level macro strategies.

### 8.5 Limitations and Future Work

**Current Limitations:**
- Weak reward shaping reduces learning quality, especially for deep models, leading to slower convergence and noisy value estimates.
- State representations omit temporal dynamics, preventing the agent from reasoning about momentum shifts or multi step strategic patterns.
- Action space mismatch across methods limits comparability, since Q-Learning uses 3 actions while DQN/DDQN operate on the full 6 action set.

**Future Improvements:**
1. Strengthen reward shaping to provide clearer, more consistent learning signals for macro decisions.
2. Incorporate sequential or recurrent state features (e.g., LSTMs or multi step histories) to model evolving early game trajectories.
3. Unify state and action abstractions across algorithms to enable fairer benchmarking and more interpretable comparisons.
4. Explore more advanced deep RL architectures—such as dueling networks, distributional RL, or multi agent extensions—to improve robustness and scalability.
